### we are going to collect to Do data

```python
import requests
import pandas as pd
```

```python
url = 'https://jsonplaceholder.typicode.com/todos'

headers = {
    "User-Agent":"Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Mobile Safari/537.36"
}

res = requests.get(url,headers=headers)
json_data = res.json()
```

```python
json_data

all_todos = []

for todo in json_data:
    
    serial_no = todo.get('id',"")
    if serial_no:
        serial_no = serial_no
    else:
        serial_no = None
    user_id = todo.get("userId","")
    if user_id:
        user_id = user_id
    else:
        user_id = None
    title = todo.get('title','')
    if title:
        title = title
    else:
        title = "NA"
    status = todo.get('completed')

    all_todos.append([serial_no,user_id,title,status])
```

```python
columns = ['Sno', 'User Id', "Tile","Status"]
df = pd.DataFrame(all_todos, columns=columns)
df.to_csv("Scraping_data/todos_data.csv",index=False)
```
