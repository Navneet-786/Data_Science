### get the data from web scraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
total_pages = 10
pages = list(range(1,total_pages + 1))

headers = {
    "User-Agent":"Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Mobile Safari/537.36"
}

all_books = []
for page in pages:
    url = f"https://books.toscrape.com/catalogue/page-{page}.html"
    res = requests.get(url, headers=headers)

    if res.status_code != 200:
        break

    soup = BeautifulSoup(res.text, 'lxml')
    articles = soup.find_all('article', class_='product_pod')

    for article in articles:
        book_name = article.select_one("h3 a").get_text(strip=True)

        price = article.select_one(".price_color").get_text(strip=True)
        price = float(price.replace("Â£", ""))

        availability = article.select_one(".instock").get_text(strip=True)

        all_books.append({
            "Book Name": book_name,
            "Price": price,
            "Status": availability
        })

df = pd.DataFrame(all_books)
```

```python
df.to_csv("Scraping_data/Book_web_page_scrapping.csv", index=False)
```
