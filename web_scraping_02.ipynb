### we going to scrap a gaming website

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
total_pages = 90
pg_lst = list(range(1, total_pages+1))

headers = {
    "User-Agent":"Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Mobile Safari/537.36"
}
all_games = []
for page in pg_lst:
    url = f"https://sandbox.oxylabs.io/products?page={page}"
    res = requests.get(url, headers=headers)

    soup = BeautifulSoup(res.text, "lxml")

    products = soup.find_all("div",class_="product-card")
    for product in products:
        #game name
       game_name = product.find("a").find("h4" ,class_="title").get_text().strip()

        #game desc
       description = product.find("p", class_="description").get_text().strip()[:400]

        #game price
       price = product.find("p", class_="description").find_next("div").get_text().replace("€", "")

        #game status
       status_tag = product.select_one("p[class*='stock']")
       status = status_tag.get_text(strip=True) if status_tag else "In Stock"

        #game categories
       tags = product.find("p", class_="category").find_all("span")
       categories = [x.get_text().strip() for x in tags ]
        
       all_games.append([game_name, description, price, categories,status])
        
       

```

```python
df = pd.DataFrame(all_games, columns=['Game','Desc',"Price","Categories","Status"])
df.to_csv("Scraping_data/scraped_game_data.csv", index=False)
df
```
