```python
#Import 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(
    data = tips,
    x = "total_bill",
    y = 'tip',
    col = "time",
    hue="sex",
    size="size"
)
```

```python
#line and scatter plot
sns.get_dataset_names()
data = sns.load_dataset('flights')


# # year vs passengers
# sns.scatterplot(
#     data = data,
#     x = "year",
#     y = "passengers" 
# 

sns.lineplot(
    data = data,
    x = "year",
    y = "passengers"
)
```

```python
#bar plot
# days vs tips acc to sex

sns.barplot(
    data = tips,
    x = "day",
    y = "tip",
    hue = "sex",
   palette="viridis"
)
```

```python
#box plot

# tips on basic of days
sns.boxplot(
    data = tips,
    x = "day",
    y = "tip",
    hue ="sex"
)
```

```python
#histogram
penguin  = sns.load_dataset('penguins')
#study of body mass vs freq
sns.histplot(
    data = penguin,
    x = "body_mass_g",
    bins = 40,
    kde=True,
     stat="density"
    # hue="sex"
)
```

```python
#heapmap
flight_data = data
pivot_data = flight_data.pivot(index="month" , columns="year" , values="passengers")
```

```python
sns.heatmap(
    data = pivot_data,
    cmap ="coolwarm",
    annot = True,
    fmt = "d"
)
plt.title("passenger in month and years")
plt.tight_layout()
```
