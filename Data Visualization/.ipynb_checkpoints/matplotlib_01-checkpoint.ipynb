### Oscar Winning Movies

```python
movies = [
    "Titanic",
    "The Lord of the Rings: The Return of the King",
    "Forrest Gump",
    "Gladiator",
    "Parasite",
    "Oppenheimer"
]

years = [1997 , 2003 , 1994 , 2000 , 2019 , 2023]
revenues = [2.26,1.15, 0.678, 0.465, 0.262, 0.952]

plt.plot(years, revenues, marker="o",color="hotpink", lw="1.5")
plt.title("Oscar Winning Movies Revenue vs year")
plt.xlabel("Years")
plt.ylabel("Revenue(in $)")
```

```python
## data is visualized not we have no info into graph that what thing tells what

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
product_A = [150, 180, 210, 190, 230, 250, 270, 260, 240, 280, 300, 320]
product_B = [130, 160, 200, 170, 210, 240, 260, 250, 220, 260, 290, 310]

plt.plot(months, product_A)
plt.plot(months, product_B)
```

```python
# this graph is better than prev as it describe everything

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
product_A = [150, 180, 210, 190, 230, 250, 270, 260, 240, 280, 300, 320]
product_B = [130, 160, 200, 170, 210, 240, 260, 250, 220, 260, 290, 310]

title_font = {"color":"hotpink"}
plt.plot(months, product_A,marker="o", label="product A")
plt.plot(months, product_B, marker="o", label="product B")
plt.title("Sales of The Years", fontdict=title_font, loc="center")
plt.xlabel("Months")
plt.ylabel("Sales(items)")
plt.grid(axis="y")
plt.legend()

```

```python
#format string
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
product_A = [150, 180, 210, 190, 230, 250, 270, 260, 240, 280, 300, 320]
product_B = [130, 160, 200, 170, 210, 240, 260, 250, 220, 260, 290, 310]

plt.plot(product_A, months,"o--r", label="Product A", )
plt.plot(product_B, months,">-g", label="Product B")
plt.legend()
plt.xlabel("Months")
plt.ylabel("Sales(items)")
plt.title("Sales of the Year")
plt.grid()
# print(plt.style.available)
# plt.style.use('default')

```
