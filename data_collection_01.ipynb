### we are going to fetch the data using API

```python
#using requests library
import requests
import pandas as pd
```

```python
URL = 'https://restcountries.com/v3.1/all?fields=name,capital,currencies'
# create a headers
headers = {
    "User-Agent":'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Mobile Safari/537.36'
}
# hit the request
res = requests.get(URL,headers=headers)
#convert into json
json_data = res.json()
```

```python
countries = []
# loop on every country
for country in json_data:
    name = country['name']['common'].strip()
    currencies = country.get("currencies",{})
    if currencies:
        currency_code = list(currencies.keys())[0].strip()
        currency_symbol = currencies[currency_code]['symbol']
        currency_name = currencies[currency_code]['name']
    else:
        currency_code = None
        currency_symbol = None
        currency_name = None
    country_capital = country.get('capital',{})
    if country_capital:
        capital = country_capital[0]
    else:
        capital = None
    countries.append([name, capital,currency_code, currency,currency_symbol])

```

```python
df = pd.DataFrame(countries, columns=['Country','Capital','Currency Code','Currency Symbol','Currency Name'])
df.to_csv('Scraping_data/country_info.csv',index=False)
```
